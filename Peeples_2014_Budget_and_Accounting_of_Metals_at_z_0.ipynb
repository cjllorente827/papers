```python

fig, ax = plt.subplots(figsize=(7,7))
baryon_frac = (0.0486)/(0.0486 + 0.2589)

sm = np.logspace(7,12)

gas_mass_peeples_2014 = lambda M: M * 10**(-0.48 * np.log10(M) + 4.39)

gm = gas_mass_peeples_2014(sm)
ax.semilogx(sm, gm)


plt.show()
```
